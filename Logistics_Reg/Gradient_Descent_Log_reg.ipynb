## SKLEARN LOGISTIC REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,random_state=41,n_classes=2,n_clusters_per_class=1,n_features=2,n_informative=1,n_redundant=0,hypercube=False,class_sep=20)
```

```python
lor = LogisticRegression(penalty=None,solver='sag')
```

```python
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
c = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.plot(x_input,y_input,color = "black",linewidth = 3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.ylim(-3,2)
```

## Using Gradient Descent from Scratch

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def gd(X,y):
    X= np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5

    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])

    return weights[0],weights[1:]
```

```python
intercept_ , coef_ = gd(X,y) 
```

```python
m = - (coef_[0]/coef_[1])
c = - (intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + c
```
